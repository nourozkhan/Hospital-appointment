```python

class Patient():

  def __init__(self, name, age, sex):

    self.name = name
    self.age = age
    self.sex = sex

  def show_doctors():
    for doctor in doctorsList:
      print('Doctor Name: ' + doctor.name)

  def show_schedule():
    for schedule in scheduleList:
      print('Schedule Time: ' + schedule.scheduleTime)
      print(schedule.scheduleDays)

  def make_appointment(doctor, patient, hospital, date, time):
    A101 = Appointment(doctor, patient, hospital, date, time)
    print('Doctor Name:' + ' ' + A101.doctor)
    print('Patient Name:' + ' ' +A101.patient)
    print('Hospital Name:' + ' ' +A101.hospital)
    print('Date Name:' + ' ' +A101.date)
    print('Time Name:' + ' ' +A101.time)
    print('Request sent for appointment please wait for confirmation')

  def cancle_appointment():
    print('appointment is cancelled')


class Doctor():

  def __init__(self, name, licenseNo, hospital):

    self.name = name
    self.licenseNo = licenseNo
    self.hospital = hospital

  def add_schedule(doctor, hospital, scheduleDays, scheduleTime):
    S101 = Schedule(doctor, hospital, scheduleDays, scheduleTime)
    print('Doctor Name:' + ' ' + S101.doctor)
    print('Hospital Name:' + ' ' +S101.hospital)
    print(S101.scheduleDays)
    print('Schedule Time:' + ' ' +S101.scheduleTime)
    print('schedule added')

  def change_schedule(self):
    print('change schedule')


class Hospital():

  def __init__(self, name, location):

    self.name = name
    self.location = location


class Schedule():

  def __init__(self, doctor, hospital, scheduleDays, scheduleTime):

    self.doctor = doctor
    self.hospital = hospital
    self.scheduleDays = scheduleDays
    self.scheduleTime = scheduleTime


class Appointment():

  def __init__(self, doctor, patient, hospital, date, time):

    self.doctor = doctor
    self.patient = patient
    self.hospital = hospital
    self.date = date
    self.time = time

  def check_appointment_length(self):
    print('Checking')


class PatientRecord():

  def __init__(self, doctor, patient, disease, dateofTreatment, treatment, medicines):

    self.doctor = doctor
    self.patient = patient
    self.disease = disease
    self.dateofTreatment = dateofTreatment
    self.treatment = treatment
    self.medicines = medicines
```
